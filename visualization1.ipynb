```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# to plot the line
plt.style.use("dark_background")
x = np.arange(10)
print(x)
y=x**2
y1=2*x+3
print(y)
print(y1)
plt.plot(x,y,color="red",label="apple",marker="o")
plt.plot(x,y1,label="kiwi",linestyle="dashed")
plt.xlabel("time")
plt.ylabel("prices")
plt.title("prices of fruits over time")
plt.legend()
plt.show()
```
